```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  # ✅ Fix this import
import librosa 
import librosa.display  # ✅ Required for visualizing MFCCs
```

```python
audio_data, sample_rate = librosa.load("songs/song1.mp3")
audio_data, sample_rate = librosa.load("songs/song2.mp3")
```

```python
def load_audio(file_path):
    """
    Loads an audio file and converts it into a waveform and sampling rate.
    :param file_path: Path to the audio file (MP3, WAV)
    :return: waveform (numpy array), sampling rate (int)
    """
    y, sr = librosa.load(file_path, sr=None)  # Load audio file with its native sampling rate
    return y, sr
```

```python
def extract_mfcc(y, sr, n_mfcc=13):
    """
    Extracts MFCC features from an audio waveform.
    :param y: Waveform array
    :param sr: Sampling rate
    :param n_mfcc: Number of MFCC features to extract (default = 13)
    :return: MFCC feature array
    """
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return mfccs
```

```python
def mfcc_to_dataframe(mfccs):
    """
    Converts MFCC feature array into a Pandas DataFrame.
    :param mfccs: MFCC feature array
    :return: DataFrame of MFCCs
    """
    mfcc_df = pd.DataFrame(mfccs)
    return mfcc_df
```

```python
def plot_mfcc(mfccs):
    """
    Plots the MFCC spectrogram for visualization.
    :param mfccs: MFCC feature array
    """
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, x_axis='time')
    plt.colorbar()
    plt.title('MFCC Spectrogram')
    plt.show()
```

```python
# ✅ Example: Extract MFCCs from a Song
file_path = "songs/song1.mp3"  # Change this to your actual file path

# Step 1: Load the song
y, sr = load_audio(file_path)

# Step 2: Extract MFCCs
mfccs = extract_mfcc(y, sr)

# Step 3: Convert MFCCs to DataFrame
mfcc_df = mfcc_to_dataframe(mfccs)

# Step 4: Plot the MFCC Spectrogram
plot_mfcc(mfccs)

# Step 5: Print the DataFrame
print(mfcc_df.head())
```
